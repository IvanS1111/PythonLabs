```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random
from scipy import interpolate
```

```python
def f1(x):
    return np.exp(2*x)*math.cos(3*x)
def f2(x):
    return np.sin(math.log(x,math.e))
def f3(x):
    return math.log(x, math.e)
def f4(x):
    return math.cos(x) + math.sin(x)
```

```python
def one_way_method_right(f, x, dx, p):
    if p == 1 :
        return (f(x + dx) - f(x))/dx
    if p == 2:
        return (f(x + 2*dx) - 2 * f(x) + f(x - 2*dx))/(2*dx)**2
    if p == 3:
        return (f(x + 2*dx) - 2 * f(x + dx) + 2 * f(x - dx) - f(x - 2*dx)) / (2*dx**3)
```

```python
def integrate(f, a, b, n):
    dx = (b - a)/(2*n)
    S = 0
    x = []
    y = []
    for i in range(n):
        S += f(a + 2*i*dx + dx)*2*dx
        x.append(a + 2*i*dx + dx)
        y.append(f(a + 2*i*dx + dx))
    #S += f(b)*dx    
    fig, ax = plt.subplots()
    ax.bar(x, y, width = 2*dx)
    fig.set_figwidth(12)   
    fig.set_figheight(6)   
    y1 = []
    for i in date:
        y1.append(f(i))
    plt.plot(date, y1, color='red')
    plt.show()
    return S
```

<h3>Метод наименьших квадратов</h3>

```python
def summ(X):
    S = 0
    for i in X:
        S += i
    return S
        
def min_kvadrat(X, Y):
    n = len(X)
    xy = 0
    for i in range(n):
        xy += X[i] * Y[i]
    xy *= n
    x_sqr = 0
    for i in range(n):
        x_sqr += (X[i])**2
    x_sqr *= n
    xy -= summ(X) * summ(Y)
    x_sqr -= (summ(X))**2
    a = xy/x_sqr
    b = (summ(Y) - a*summ(X))/n
    return (a, b)

a, b = min_kvadrat(date, cases)
fig, ax = plt.subplots()
fig.set_figwidth(12)   
fig.set_figheight(12)   
plt.plot(date, cases)
interplt = []
for i in date:
    interplt.append(a * i + b)
plt.plot(date, interplt)
```
